```python
%%sql
-- Identify matches played between two specific teams (e.g., India and South Africa) in 2024 and their results.

SELECT Team1, Team2, Winner, MatchDate

    FROM T20I
WHERE 
((Team1 = 'India' and Team2 = 'South Africa') or (Team1 = 'South Africa' and Team2 = 'India'))
and MatchDate LIKE '%2024' ;
```

```python
%%sql
-- Find the team with the highest number of wins in 2024 and the total matches it won.

SELECT Winner , COUNT(*) AS ' Number of wins' 
From T20I
WHERE MatchDate LIKE '%2024'
GROUP BY Winner
ORDER BY Count(*) Desc LIMIT 1 
;
```

```python
%%sql

-- Rank the teams based on the total number of wins in 2024.

WITH RANK_TEAM AS(
SELECT Winner , COUNT(*) AS 'Total_win'

FROM T20I
WHERE MatchDate LIKE '%2024'
GROUP BY Winner)

SELECT Winner, Total_--Q4 Which team had the highest average winning margin (in runs), and what was the average margin?
win ,
RANK () OVER (
    ORDER BY Total_win DESC) AS 'Rank'
FROM RANK_TEAM

;
```

```python
%%sql
-- Which team had the highest average winning margin (in runs), and what was the average margin?
SELECT Winner,
    AVG(
        CAST(
            SUBSTR(Margin, 1, INSTR(Margin,' ') - 1)
            AS INTEGER
        )
    ) AS Avg_Margin
FROM T20I
WHERE Margin LIKE '%run%'
GROUP BY Winner
ORDER BY Avg_Margin DESC
LIMIT 1;
```

```python
%%sql
-- Which team had the highest average winning margin (in wickets), and what was the average margin?
SELECT Winner , AVG(
    CAST(
        SUBSTR(Margin, 1, INSTR(Margin,' ') -1)
        AS INTEGER)
) AS Avg_Margin
FROM T20I
WHERE Margin LIKE '%wicket%'
GROUP BY Winner
ORDER BY Avg_Margin DESC
LIMIT 1;
```

```python
%%sql
-- List all matches where the winning margin was greater than the average margin across all matches.

WITH MATCHES AS(
SELECT Team1, Team2, Winner, Margin , MatchDate, 
    CAST( 
        SUBSTR(Margin , 1 , INSTR(Margin, ' ') -1) as Integer) as 'margin_number'
    
FROM T20I
WHERE Margin LIKE '%runs'
) ,

AVG_MARG AS(
SELECT AVG(margin_number) AS 'avg_margin'
FROM MATCHES 
)

SELECT Team1, Team2, Winner, MatchDate, margin_number , avg_margin

FROM AVG_MARG , MATCHES
WHERE  margin_number > avg_margin
ORDER BY margin_number DESC
;
```

```python
%%sql
--Find the team with the most wins when chasing a target (wins by wickets)

SELECT COUNT(*)as 'number_of_wins' , Winner, Margin
FROM T20I
WHERE Margin LIKE '%wickets'
GROUP BY Winner
ORDER BY number_of_wins DESC
LIMIT 1

;
```

```python
%%sql
--Head-to-head record between two selected teams (e.g., England vs Australia).

SELECT Winner, COUNT(*) AS Wins 
FROM T20I
WHERE Team1 LIKE '%England' and Team2 LIKE '%Australia' 
OR Team2 LIKE '%England' and Team1 LIKE '%Australia' 
GROUP BY Winner

;
```

```python
%%sql
-- Identify the month in 2024 with the highest number of T20I matches played.

SELECT 
    SUBSTR(MatchDate, 1, 3) AS MonthName,
    COUNT(*) AS MatchesPlayed

FROM T20I
WHERE MatchDate LIKE '%2024'
GROUP BY MonthName
ORDER BY MatchesPlayed DESC
;
```

```python
%%sql
-- For each team, find how many matches they played in 2024 and their win percentage.

WITH MatchPlayed as(
    SELECT Team1 AS Team
    FROM T20I
    WHERE MatchDate LIKE '%2024'
    UNION ALL
    SELECT Team2 AS Team
    FROM T20I
    WHERE MatchDate LIKE '%2024'
),

    Played as(
        SELECT Team, COUNT(*) AS MatchNumber
                FROM MatchPlayed
                GROUP BY Team),
                           
    Wins as(
       SELECT Winner as Team, COUNT(*) AS WinNumber
              FROM T20I
        Group BY Winner)

            
    
SELECT 
        Played.Team, MatchNumber , WinNumber,
        ROUND( WinNumber *100 / MatchNumber, 2) as WinPercentage
 FROM Played
 LEFT JOIN Wins

ON Played.Team = Wins.Team
        ORDER BY WinPercentage DESC;   
```

```python
%%sql
-- Identify the most successful team at each ground (team with most wins per ground).
WITH Wins as (
    SELECT Winner, Ground , COUNT(*) as WinNumber

    FROM T20I
    GROUP BY ground, Winner),
                                 
Ranked as (
    SELECT Winner, Ground, WinNumber ,
    Rank() over (Partition BY Ground ORDER  BY WinNumber DESC) as WinRank
    from Wins
    )

SELECT *
FROM Ranked
WHERE WinRank =1
;

```
